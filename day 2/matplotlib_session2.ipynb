```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# colored scatterplots

iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
iris['Species'] = iris['Species'].replace('Iris-setosa', 0)
iris['Species'] = iris['Species'].replace('Iris-versicolor', 1)
iris['Species'] = iris['Species'].replace('Iris-virginica', 2)
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
# increase size
plt.figure(figsize=(12, 6))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()

```
