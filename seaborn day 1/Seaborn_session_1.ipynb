```python
# 2types of functions
# 1. Figure level function
# 2. Axes level function

# 1. relational plot
# to see the statistical relation between 2 or more variables.
# bivariate analysis

# Scatterplot
# Lineplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatterplot -> Axes level function

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# figure level function -> recommended is figure level function
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# default kind is 'scatter'
sns.relplot(x='total_bill', y='tip', data=tips, kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time', size='size')
plt.show()
```

```python
# line plot

gap = px.data.gapminder()
gap

temp_df = gap[gap['country'] == 'Nepal']
temp_df
```

```python
# axes level function
sns.lineplot(x='year', y='lifeExp', data=temp_df)
plt.show()
```

```python
# axes level function
sns.lineplot(x='year', y='pop', data=temp_df)
plt.show()
```

```python
# using relplot -> figure level function
sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line') # default kind is 'scatter'
plt.show()
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['Nepal', 'India', 'China'])]
sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line', hue='country', style='country')
plt.show()
```

```python
# facet plot -> only works with fig level function, relplot
# Does not work with scatterplot and lineplot, axes level function

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='sex')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker', row='sex')
plt.show()
```

```python
# col wrap
gap
sns.relplot(x='lifeExp', y='gdpPercap', data=gap, kind='line', col='continent', col_wrap=2)
plt.show()
#
```

```python
# 2. Distribution plot
# histplot, kdeplot, & rugplot

# used for univariate analysis
# used to find out distribution
# range of the observation
# central tendency
# is the data bimodal?
# are there outliers?

# figure level -> displot
# axex level -> histplot, kdeplot, rugplot
```

```python
sns.histplot(tips['total_bill'])
plt.show()
```

```python
sns.histplot(x = tips['total_bill'])
plt.show()
```

```python
sns.histplot(x = tips['total_bill'], data=tips)
plt.show()
```

```python
# kind = 'hist' is default here
sns.displot(x = tips['total_bill'], data=tips)
plt.show()
```

```python
sns.displot(x = tips['total_bill'], data=tips, bins = 2)
plt.show()
```

```python
# it's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# also called countplot
sns.displot(x = tips['sex'], data=tips)
plt.show()
```

```python
# hue
sns.displot(x = tips['sex'], data=tips, hue='sex')
plt.show()
```

```python
sns.displot(x = tips['sex'], data=tips, hue='sex', element='step')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(x = titanic['age'], data=titanic, element='step', hue='sex')
plt.show()
```
